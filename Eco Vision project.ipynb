## CHAPTER 1 - AIR POLLUTION

### AIR POLLUTANTS AND TRANSPORTATION

Transportation, especially road vehicles, is a major contributor to air pollution in Europe and all over the World.


The combustion of fossil fuels in vehicles' engines releases a variety of pollutants into the atmosphere. Overall, road vehicles generate air pollution primarily through the combustion of fossil fuels in their engines, emitting pollutants such as nitrogen oxides, particulate matter, volatile organic compounds, carbon monoxide, and carbon dioxide. Addressing vehicle emissions through cleaner fuels, advanced engine technologies, and stringent emissions standards is essential for improving air quality and public health.
Analyzing in detail the pollutant emissions of vehicles we can distinguish these:
* **Combustion of Fossil Fuels**: most road vehicles run on gasoline or
diesel, both of which are derived from fossil fuels (when these fuels are burned in internal combustion engines, they produce emissions that contribute to air pollution); gasoline-powered vehicles emit pollutants such as carbon monoxide (CO), nitrogen oxides (NOx), volatile organic compounds (VOCs), and particulate matter (PM); diesel-powered vehicles are known for emitting higher levels of nitrogen oxides and particulate matter compared to gasoline vehicles (diesel engines produce larger quantities of NOx due to their higher combustion temperatures, and they emit more fine particulate matter (PM2.5) and black carbon): for this reason the European Commission has decided that after 2035 new diesel-powered vehicles won’t be sallable.
* **Nitrogen Oxides (NOx)**: nitrogen oxides, including nitrogen dioxide (NO2) and nitrogen monoxide (NO), are produced during the combustion process in vehicle engines and contribute to the formation of ground-level ozone (O3) and secondary particulate matter, both of which have adverse effects on human health and the environment;
Particulate Matter (PM): is a complex mixture of tiny particles and liquid droplets suspended in the air; vehicles emit both primary particulate matter (directly emitted from tailpipes) and secondary particulate matter (formed in the atmosphere through chemical reactions).
* **Volatile Organic Compounds (VOCs)**: are organic chemicals that evaporate readily into the air at room temperature and contribute to the formation of ground-level ozone (O3) and secondary organic aerosols, which can exacerbate respiratory issues and contribute to smog formation; vehicle emissions contain VOCs from fuel evaporation and incomplete combustion processes.
* Carbon Monoxide (CO) and Carbon Dioxide (CO2): while modern vehicles have largely reduced CO emissions through improved engine technology and catalytic converters, it remains a concern in older vehicles and in areas with heavy traffic congestion; carbon dioxide is a greenhouse gas emitted in large quantities by road vehicles: although it is not directly toxic to human health like other pollutants, CO2 is a major contributor to climate change and global warming;
* **Emissions Standards and Control Technologies**: To address the air pollution from road vehicles, European countries have implemented stringent emissions standards, such as the Euro standards, which set limits on the emissions of pollutants like NOx, PM, and CO; vehicle manufacturers employ various technologies to meet these standards, including catalytic converters, diesel particulate filters (DPF), selective catalytic reduction (SCR) systems for NOx reduction, and gasoline direct injection (GDI) systems for improved fuel efficiency and reduced emissions.

We must pay attention to the column measure: CO2 and GHG are different from the others gas emissions; we will consider two groups of measure: KG_CAP together TONNE_CAP (emissions per capita) and THND_TONNE together THND_TONNECO2 and MLN_TONNE.

At first will focus on air pollution related to World countries, then we will analyze the European countries. First of all we will clean country pollution dataset from NaN values; here we filter the datas in THND_TONNE together THND_TONNECO2 and MLN_TONNE:

```python
# Conversion factors to convert various measurements to thousand tonnes (THND_TONNE)
conversion_factors = {
   #"MLN_TONNE": 1e6,  # Convert million tonnes to thousand tonnes
   #"THND_TONNE": 1,   # No conversion needed for thousand tonnes
   #"THND_TONNECO2": 1,  # No conversion needed for THND_TONNECO2
    "THND_TONNECO2": 1e-3,  # Convert thousand tonnes of CO2 to million tonnes
    "THND_TONNE": 1e-3,   # Convert thousand tonnes to million tonnes
    "MLN_TONNE": 1   # No conversion needed for million tonnes
    # Add more conversion factors as needed
}
# df1 contains only some columns of the file COUNTRIES_POLLUTION
df1=pd.DataFrame(COUNTRIES_POLLUTION,columns=['LOCATION', 'SUBJECT', 'MEASURE', 'TIME', 'Value'])

# Define the list of valid measures
valid_measures = ["MLN_TONNE", "THND_TONNE", "THND_TONNECO2"]

# df2 contains only rows with measure "THND_TONNE", "MLN_TONNE", "THND_TONNECO2"
# Filter the DataFrame based on the "measure" column
df2 = df1[df1['MEASURE'].isin(valid_measures)].copy()

# Function to standardize values to thousand tonnes
def standardize_to_thousand_tonnes(value, measure):
    conversion_factor = conversion_factors.get(measure, None)
    if conversion_factor is None:
        raise ValueError(f"No conversion factor found for measure: {measure}")
    return value * conversion_factor

# Apply standardization function to each row
df2["Value_MLN_TONNE"] = df2.apply(lambda row: standardize_to_thousand_tonnes(row["Value"], row["MEASURE"]), axis=1)

# countries_pollution doesn't contains Nan values and we filter only years between 1990 and 2021
countries_pollution = df2.dropna(subset=["Value"])
countries_pollution=countries_pollution[(countries_pollution["TIME"]>=1990) &  (countries_pollution["TIME"]<=2021)]

# now we replace the alpha-3 country codes in the 'LOCATION' column of the DataFrame with the corresponding country names
# using the mapping dictionary country_mapping_alpha_3:
countries_pollution['LOCATION']=countries_pollution['LOCATION'].map(country_mapping_alpha_3)
print(countries_pollution)
```

Now we consider the datas with KG_CAP together with TONNE_CAP:

```python
# Conversion factors to convert various measurements to tonnes (TONNE_CAP)
# Function to standardize values from KG_CAP to TONNE_CAP

# df0 contains only some columns of the file COUNTRIES_POLLUTION
df0=pd.DataFrame(COUNTRIES_POLLUTION,columns=['LOCATION', 'SUBJECT', 'MEASURE', 'TIME', 'Value'])
# Define the list of valid measures
valid_measures2 = ["TONNE_CAP", "KG_CAP"]

# df contains only rows with measure "TONNE_CAP", "KG_CAP"
# Filter the DataFrame based on the "measure" column
df = df0[df0['MEASURE'].isin(valid_measures2)].copy()
def standardize_to_tonne_cap(row):
    if row["MEASURE"] == "KG_CAP":
        return row["Value"] / 1000  # Convert kilograms to tonnes
    else:
        return row["Value"]  # No conversion needed for TONNE_CAP

# Apply standardization function to each row
df["Value_TONNE_CAP"] = df.apply(standardize_to_tonne_cap, axis=1)

# Print the DataFrame with standardized values
print(df)

# countries_pollution doesn't cotains Nan values and we filter only years between 1990 and 2021
countries_pollutionCAP = df.dropna(subset=["Value"])
countries_pollutionCAP=countries_pollutionCAP[(countries_pollutionCAP["TIME"]>=1990) &  (countries_pollutionCAP["TIME"]<=2021)]

# now we replace the alpha-3 country codes in the 'LOCATION' column of the DataFrame with the corresponding country names
# using the mapping dictionary country_mapping_alpha_3:
countries_pollutionCAP['LOCATION']=countries_pollutionCAP['LOCATION'].map(country_mapping_alpha_3)
print(countries_pollutionCAP)
```

European countries have implemented various policies and regulations aimed at reducing air pollution from vehicles and improving air quality. We can detail some of these policies:


1.   **Emissions Standards (Euro Standards)**: the European Union (EU) has implemented a series of emissions standards known as the Euro standards, which set limits on the emissions of pollutants from vehicles and specify maximum allowable limits for pollutants such as nitrogen oxides (NOx), particulate matter (PM), carbon monoxide (CO), and hydrocarbons (HC) for different vehicle categories (e.g., passenger cars, trucks, buses); the standards have become progressively more stringent over time, with newer Euro standards imposing lower emission limits. For example, Euro 1 was introduced in 1992, and subsequent standards (Euro 2, Euro 3, etc.) have been implemented at intervals to further reduce vehicle emissions;
2.   **Low Emission Zones (LEZs) and Congestion Charging**: many European cities have established Low Emission Zones (LEZs) where only vehicles meeting specific emissions standards are allowed to enter and which impose restrictions on the types of vehicles (e.g., diesel-powered) and their emission levels, aiming to reduce air pollution in urban areas; some cities have also implemented congestion charging schemes, where vehicles are charged for entering designated areas during peak hours, encouraging the use of public transportation and reducing traffic congestion and emissions;
3.   **Vehicle Scrappage Programs**: several European countries have implemented vehicle scrappage programs having the aim of removing older and more polluting vehicles from the road; these programs typically offer financial incentives or subsidies to encourage owners to replace their old vehicles with newer, cleaner models that comply with current emissions standards, thus helping reduce emissions and improve air quality;
4.   **Promotion of Electric and Alternative Fuel Vehicles**: European countries have introduced various incentives and support measures to promote the adoption of electric and alternative fuel vehicles (AFVs), such as hybrid and hydrogen-powered vehicles (such as tax breaks, subsidies for vehicle purchase, and grants for installing charging infrastructure); additionally, EU regulations have set targets for reducing CO2 emissions from new vehicles, incentivizing automakers to produce more fuel-efficient and low-emission vehicles;
5.   **Vehicle Inspection and Maintenance Programs**: European countries have established vehicle inspection and maintenance programs to ensure that vehicles meet emissions standards and are properly maintained, inspections that help identify and address issues such as malfunctioning emission control systems and excessive emissions from older vehicles (vehicles failing to meet emission standards may be required to undergo repairs or upgrades to reduce emissions or face penalties);
6.   **Promotion of Sustainable Transportation Modes**: European countries have implemented policies to promote sustainable transportation modes such as cycling, walking, and public transportation; investments in cycling infrastructure, pedestrian zones, and public transit networks aim to reduce reliance on private vehicles and decrease emissions from transportation.

These policies and measures demonstrate European countries' efforts to combat air pollution from vehicles and promote cleaner and more sustainable transportation options. Continued implementation and enforcement of these policies are essential for achieving further reductions in air pollution and improving public health and environmental quality.
